### 2. 리스트의 합

```python
def Sum(lst):
    s = 0
    for num in lst:
        s += num
    return s

t = int(input())
for _ in range(t):
    lst = list(map(int, input().split()))
    print(Sum(lst))
```

### 3. 스택

```python
import collections

t = int(input())
for _ in range(t):
    stack = []
    ans = []
    qry = list(map(int, input().split()))
    for q in qry:
        if q > 0:
            stack.append(q)   # push 역할
        else:
            
            ans.append(stack.pop())    # pop 역할
    print(*ans)
```

### 4. 큐

```python
import collections

t = int(input())
for _ in range(t):
    queue = []
    ans = []
    qry = list(map(int, input().split()))
    for q in qry:
        if q > 0:
            queue.append(q)         # enqueue 역할: 큐의 뒤에 값 추가
        else:
            if queue:               # 큐가 비어있지 않을 때만 dequeue 수행
                dequeued = queue.pop(0)  # dequeue 역할: 큐의 맨 앞의 값 제거
                ans.append(dequeued)
    print(*ans)
```

### 5. 우선순위 큐

```python
import heapq

t = int(input())
for _ in range(t):
    n = int(input())
    cmds = list(map(int, input().split()))
    hq = []
    result = []

    for cmd in cmds:
        if cmd > 0:
            heapq.heappush(hq, cmd)
        elif cmd == -1 and hq:
            result.append(str(heapq.heappop(hq)))

    print(" ".join(result))
```

### 6. 괄호 검사

```python
def is_balanced(brackets):
    stack = []
    pair = {')': '(', ']': '[', '}': '{'}

    for ch in brackets:
        if ch in '([{':
            stack.append(ch)
        elif ch in ')]}':
            if not stack or stack[-1] != pair[ch]:
                return "NO"
            stack.pop()
    return "YES" if not stack else "NO"


t = int(input())
for _ in range(t):
    s = input().strip()
    print(is_balanced(s))
```

### 7. 가장 가파른 두 점

```python
t = int(input())
for _ in range(t):
    n = int(input())
    points = []
    
    while len(points) < n:
        line = input().strip()
        if not line:
            continue
        nums = list(map(int, line.split()))
        for i in range(0, len(nums), 2):
            if i+1 < len(nums):
                points.append((nums[i], nums[i+1]))

    points.sort()  # x좌표 기준 정렬

    max_s = -1
    ans = ()

    for i in range(n - 1):
        (x1, y1), (x2, y2) = points[i], points[i+1]
        s = abs(y1 - y2) / abs(x1 - x2)

        if s > max_s:
            max_s = s
            ans = (x1, y1, x2, y2)
        elif s == max_s:
            if (x1, y1, x2, y2) < ans:
                ans = (x1, y1, x2, y2)

    print(*ans)
```

### 1. 피보나치 수열

```python
def fibonacci(n):
    if n == 1 or n == 2:
        return 1
    fib = [0] * (n+1)
    fib[1], fib[2] = 1, 1
    for i in range(3, n+1):
        fib[i] = fib[i-1] + fib[i-2]
    return fib[n]

# 입력 처리
t = int(input())  # 테스트 케이스 수
for _ in range(t):
    n = int(input())
    print(fibonacci(n))
```

### 2. 이진탐색

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

t = int(input())  # 테스트 케이스 수

for _ in range(t):
    arr = list(map(int, input().split()))      # 정렬된 리스트
    targets = list(map(int, input().split()))  # 찾을 숫자 리스트

    result = [binary_search(arr, x) for x in targets]
    print(*result)
```

### 3. 하노이 탑

```python
def hanoi(n, start, end, via, moves):
    if n == 1:
        moves.append(f"{start} -> {end}")
    else:
        hanoi(n-1, start, via, end, moves)   # 1단계
        moves.append(f"{start} -> {end}")    # 2단계
        hanoi(n-1, via, end, start, moves)   # 3단계

t = int(input())
for _ in range(t):
    n = int(input())
    moves = []
    hanoi(n, 'A', 'C', 'B', moves)
    for move in moves:
        print(move)
```

### 4. 가장 가까운 두 점

```python
def manhattan(p1, p2):
    return abs(p1[0] - p2[0]) + abs(p1[1] - p2[1])

t = int(input())
for _ in range(t):
    n = int(input())
    points = [tuple(map(int, input().split())) for _ in range(n)]

    min_dist = float('inf')
    for i in range(n):
        for j in range(i+1, n):
            dist = manhattan(points[i], points[j])
            if dist < min_dist:
                min_dist = dist

    print(min_dist)
```

### 3. DFS

```python
import sys
sys.setrecursionlimit(1000000)

def dfs(u, adj, visited, result):
    visited[u] = True
    result.append(u)
    for v in adj[u]:
        if not visited[v]:
            dfs(v, adj, visited, result)

T = int(input())
for _ in range(T):
    N, M = map(int, input().split())
    adj = [[] for _ in range(N)]

    for _ in range(M):
        u, v = map(int, input().split())
        adj[u].append(v)
        adj[v].append(u)

    # 번호가 작은 정점을 먼저 탐색하기 위해 인접 리스트 정렬
    for neighbors in adj:
        neighbors.sort()

    visited = [False] * N
    result = []
    dfs(0, adj, visited, result)

    print(' '.join(map(str, result)))
```

### 4. BFS

```python
from collections import deque

T = int(input())
for _ in range(T):
    N, M = map(int, input().split())
    graph = [[] for _ in range(N)]

    for _ in range(M):
        u, v = map(int, input().split())
        graph[u].append(v)

    # 오름차순 방문을 위해 인접 리스트 정렬
    for adj in graph:
        adj.sort()

    visited = [False] * N
    result = []
    queue = deque([0])
    visited[0] = True

    while queue:
        u = queue.popleft()
        result.append(u)
        for v in graph[u]:
            if not visited[v]:
                visited[v] = True
                queue.append(v)

    print(' '.join(map(str, result)))
```

### 5. 최단 경로 구하기

```python
import heapq
import sys

INF = float('inf')

def dijkstra(start, graph, N):
    dist = [INF] * N
    dist[start] = 0
    pq = [(0, start)]  # (비용, 정점)

    while pq:
        cost, u = heapq.heappop(pq)
        if cost > dist[u]:
            continue
        for v, w in graph[u]:
            if dist[v] > cost + w:
                dist[v] = cost + w
                heapq.heappush(pq, (dist[v], v))
    return dist

T = int(input())
for _ in range(T):
    N, M = map(int, input().split())
    graph = [[] for _ in range(N)]

    for _ in range(M):
        u, v, c = map(int, input().split())
        graph[u].append((v, c))  # 방향성 그래프

    dist = dijkstra(0, graph, N)
    print(dist[N-1] if dist[N-1] != INF else -1)
```

### 6. 위상 정렬

```python
from collections import deque

T = int(input())
for _ in range(T):
    N, M = map(int, input().split())
    graph = [[] for _ in range(N)]
    in_degree = [0] * N

    for _ in range(M):
        u, v = map(int, input().split())
        graph[u].append(v)
        in_degree[v] += 1

    # 진입차수 0인 노드를 오름차순으로 큐에 추가
    queue = deque()
    for i in range(N):
        if in_degree[i] == 0:
            queue.append(i)

    result = []
    while queue:
        # 오름차순으로 방문하기 위해 정렬된 상태에서 하나 꺼내기
        current = min(queue)
        queue.remove(current)
        result.append(current)
        for neighbor in graph[current]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)

    if len(result) == N:
        print(' '.join(map(str, result)))
    else:
        print(-1)
```

### 7. 가장 짧은 사이클

```python
from collections import deque

def shortest_cycle(N, graph):
    min_cycle = float('inf')

    for start in range(N):
        dist = [-1] * N
        prev = [-1] * N
        queue = deque()
        queue.append(start)
        dist[start] = 0

        while queue:
            u = queue.popleft()
            for v in graph[u]:
                if dist[v] == -1:
                    dist[v] = dist[u] + 1
                    prev[v] = u
                    queue.append(v)
                elif prev[u] != v:  # 사이클 발견 (역방향 제외)
                    cycle_len = dist[u] + dist[v] + 1
                    min_cycle = min(min_cycle, cycle_len)

    return min_cycle if min_cycle != float('inf') else -1

# 입력 처리
T = int(input())
for _ in range(T):
    N, M = map(int, input().split())
    graph = [[] for _ in range(N)]

    for _ in range(M):
        u, v = map(int, input().split())
        graph[u].append(v)
        graph[v].append(u)

    print(shortest_cycle(N, graph))
```

### 8. 아름다운 산책로

```python
def find(parent, x):
    if parent[x] != x:
        parent[x] = find(parent, parent[x])
    return parent[x]

def union(parent, a, b):
    a = find(parent, a)
    b = find(parent, b)
    if a == b:
        return False
    parent[b] = a
    return True

T = int(input())
for _ in range(T):
    N, M = map(int, input().split())
    edges = []
    for _ in range(M):
        u, v, w = map(int, input().split())
        edges.append((w, u, v))  # 가중치 기준 정렬을 위해 먼저

    edges.sort(reverse=True)  # 최대 아름다움 → 내림차순

    parent = list(range(N))
    total = 0
    count = 0

    for w, u, v in edges:
        if union(parent, u, v):
            total += w
            count += 1
            if count == N - 1:
                break

    print(total)
```
