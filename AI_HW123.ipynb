# 1. 두 바퀴 레이스

```python
def is_order_changed(seq):
    pos = {}
    start_order = []
    end_order = []

    for idx, num in enumerate(seq):
        if num not in pos:
            pos[num] = len(start_order)
            start_order.append(num)
        else:
            end_order.append(num)

    # 출발 순서 기준 인덱스를 도착 순서 순서대로 나열한 배열 생성
    mapped = [pos[num] for num in end_order]

    # 이 mapped 배열이 정렬되어 있지 않으면 순서가 바뀐 것
    for i in range(1, len(mapped)):
        if mapped[i-1] > mapped[i]:
            return "YES"
    return "NO"

# 입력 처리
t = int(input())
for _ in range(t):
    seq = list(map(int, input().split()))
    print(is_order_changed(seq))
    
# 리스트로 하면 시간제한 걸리니 큐로 풀기
```

# 2. 이진탐색2

```python
import bisect

def find_closest(sorted_list, target):
    idx = bisect.bisect_left(sorted_list, target)
    candidates = []
    
    if idx < len(sorted_list):
        candidates.append(sorted_list[idx])
    if idx > 0:
        candidates.append(sorted_list[idx - 1])
    
    # 절댓값 차이 기준으로 최소값, 같으면 작은 값 선택
    candidates.sort(key=lambda x: (abs(x - target), x))
    return candidates[0]

# 입력 처리
t = int(input())

for _ in range(t):
    base = list(map(int, input().split()))
    targets = list(map(int, input().split()))
    result = [find_closest(base, target) for target in targets]
    print(*result)
```

# 3. 해커

```python
from collections import defaultdict, deque

def count_components(n, edges):
    graph = defaultdict(list)
    for a, b in edges:
        graph[a].append(b)
        graph[b].append(a)

    visited = [False] * n
    count = 0

    def bfs(start):
        queue = deque([start])
        visited[start] = True
        while queue:
            node = queue.popleft()
            for neighbor in graph[node]:
                if not visited[neighbor]:
                    visited[neighbor] = True
                    queue.append(neighbor)

    for i in range(n):
        if not visited[i]:
            bfs(i)
            count += 1

    return count

# 입력 처리
t = int(input())
for _ in range(t):
    n, m = map(int, input().split())
    edges = [tuple(map(int, input().split())) for _ in range(m)]
    print(count_components(n, edges))
```
