# 실습4 - 가변 매개변수

```python
def vector_sum(vector, *vectors): #나머지 변수는 * 별을 붙여주기!!
    result = [vector[0], vector[1]]  #Shallow copy 방지를 위해

    for v in vectors:
        result[0] += v[0]
        result[1] += v[1]

    return result

    
v1 = [0,1]
v2 = [0.5, 0.5]
v3 = [1,0]
v4 = [6,4]
v5 = [3.13, 2.72]
m1 = vector_sum(v1, v2, v3)
m2 = vector_sum(v1, v2, v3, v4)
m3 = vector_sum(v3, v5)
print(m1,m2,m3)
```

# 실습9

```python
def comb_dict(d1, d2):
    d = d1  #shallow copy 발생 그래서 내장모듈 copy를 사용하여 deep copy해야함

    for k, v in d2.items():
        if k in d:
            d[k] += v
        else:
            d[k] = v
    return d


d1 = {'a': 1, 'b': 3, 'd': 7, 'e': 8}
d2 = {'a': 2, 'c': 4, 'e': 1}
d3 = comb_dict(d1, d2)
print(d3)
print(d1)  #Shallow copy때문에 덮어쓰기가 됨
```

```python
import copy

def comb_dict(d1, d2):
    d = copy.deepcopy(d1)  #shallow copy 발생 그래서 내장모듈 copy를 사용하여 deep copy해야함

    for k, v in d2.items():
        if k in d:
            d[k] += v
        else:
            d[k] = v
    return d


d1 = {'a': 1, 'b': 3, 'd': 7, 'e': 8}
d2 = {'a': 2, 'c': 4, 'e': 1}
d3 = comb_dict(d1, d2)
print(d3)
print(d1)  #Shallow copy때문에 덮어쓰기 안됨
```

# 실습2

```python
# 1) score.txt(읽기모드), report.txt(쓰기모드) 열기
f_score = open("score.txt","r")
f_report = open("report.txt","w")

# 2) score.txt 파일 읽기
for line in f_score:

    # 3) 평균 계산
    sid, mid, fin = line.split() #문자열
    avg = (int(mid+fin))/2
    
    # 4) 힉잠 계산
    grade = 'A'

    if avg >= 90:
        grade = 'A'
    elif avg >= 80:
        grade = 'B'
    elif avg >= 70:
        grade = 'C'  
    elif avg >= 60:
        grade = 'D' 
    else:
        grade = 'F'

        
    # 5) report.txt.에 적을 문자열 만들기
    data = sid + "  " + str(avg) + "(" + grade + ") \n"
    #print(data)

    # 6) report.txt.에 쓰기
    f_report.write(data)


f_score.close()
f_report.close()
```

```python
class Building:
    def __init__(self, n ='공학1동', h = 3, e = True):
        # self <- 호출한 instance
        self.name = n
        self.h = h
        self.elev = e

    def get_info(self):
        print("건물 이름=", self.name)
        print("건물 층수=", self.h)
        print("엘레베이터 유무=", self.elev)

a = Building()
b = Building("공학2동", 10, True)

a.get_info()  # 첫번째 인자 <- a / self <- a
print()
b.get_info()  # 첫번째 인자 <- b / self <- b
```

# 클래스 실습

```python
class Person:
    def __init__(self,n,a,h,w):
        # 성명, 나이, 키, 몸무게 초기화
        self.name = n
        self.age = a
        self.height = h
        self.weight = w

    def add_age(self):
        # self의 나이를 1증가
        self.age += 1
    
    def change_height(self, h):
        # self의 키를 h로 세팅
        self.height = h
        
    def change_weight(self, w):
        # self의 키를 h로 세팅
        self.weight = w
        
    def get_info(self):
        print("이름: ", self.name)
        print("나이: ", self.age)
        print("키: ", self.height)
        print("몸무게: ", self.weight)

a = Person("홍길동",10,100,20) 
a.get_info()

print()

a.add_age()
a.get_info()

print()

a.change_weight(40)
a.get_info()
```
