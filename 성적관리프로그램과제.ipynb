# 성적 관리 프로그램

```python
#(문제)
#파일로부터 데이터를 읽어서 성적 목록을 만들어 관리하는 성적 관리 프로그램을 작성한다.
```

```python
# 1) score.txt(읽기모드), report.txt(쓰기모드) 열기
f_score = open("score.txt","r")
f_report = open("report.txt","w")

# 2) score.txt 파일 읽기
for line in f_score:

    # 3) 평균 계산
    sid, mid, fin = line.split() #문자열
    avg = (int(mid+fin))/2
    
    # 4) 힉잠 계산
    grade = 'A'

    if avg >= 90:
        grade = 'A'
    elif avg >= 80:
        grade = 'B'
    elif avg >= 70:
        grade = 'C'  
    elif avg >= 60:
        grade = 'D' 
    else:
        grade = 'F'

        
    # 5) report.txt.에 적을 문자열 만들기
    data = sid + "  " + str(avg) + "(" + grade + ") \n"
    print(data)

    # 6) report.txt.에 쓰기
    f_report.write(data)


f_score.close()
f_report.close()
```
